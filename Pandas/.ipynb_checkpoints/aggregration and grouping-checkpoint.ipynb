```python
import pandas as pd

df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})  
```

```python
df
```

```python
df[["Department","Salary"]]
```

```python
df["Salary"].mean()
```

```python
df.groupby("Department")["Salary"].agg(["mean","max","min"])
```

```python
df.groupby("Team").filter(lambda x : x["Salary"].max() > 89)
```

```python
df.groupby("Gender")["Salary"].count()
```

```python
df.groupby("Department").filter(lambda x: x["Salary"].min() > 80)
```

```python
df["Team Avg"] = df.groupby("Team")["Salary"].transform("mean")
```

```python
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})

departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
```

```python
employees
```

```python
departments
```

```python
pd.merge(employees,departments,on="DeptID",how="left")
```

```python
pd.merge(employees,departments, on="DeptID", how="outer")
```
