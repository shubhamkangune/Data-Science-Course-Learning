```python
import numpy as np
kohli = [200, 300, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [100, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 2200]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
# plt.plot(years,kohli, color="red",label="Kohli")
# plt.plot(years,sehwag, color="black",label="Sehwag")

x = np.arange(len(years))
width = 0.4
plt.bar(x-width/2,kohli,width=width, label="kohli")
plt.bar(x+width/2,sehwag,width=width, label="sehwag")
plt.legend()
plt.show()
```

```python

plt.plot(years,kohli, label="Virat kohli")
plt.plot(years,sehwag, label="Virendar sehwag")
plt.xlabel("Years")
plt.ylabel("Runs scored")
plt.title("Cricket score yearly")
plt.legend()
plt.show()
```

```python
plt.plot(years, kohli, color="orange",linestyle=":",linewidth=6 ,label="virat")
plt.plot(years,sehwag,color="pink",linestyle=":",label="sehwag")
plt.tight_layout()
plt.grid(True)
plt.style.use("seaborn-v0_8-bright")
```
