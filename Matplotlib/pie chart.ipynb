```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
# Data
colors = ["#FC9583","#57FFF5","#C2B64F","#89FF82"]
labels = ["Sachin", "Water", "Kohli", "Yuvraj"]
runs = [18000, 80000, 12000, 5100]
explode = [0.5,0.2,0.1,0.2]
plt.pie(runs,
        labels=labels,
        colors=colors,
        explode=explode,
        shadow=True,
        startangle=280,
        autopct='%1.1f%%',
       wedgeprops={'edgecolor':'black',"linestyle":'-','linewidth':2,'fill':True})
plt.show()
```
